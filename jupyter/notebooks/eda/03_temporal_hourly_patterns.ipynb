```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

df = pd.read_csv("../../data/raw/2020-Apr-L.csv")
```

```python
df["event_time"] = pd.to_datetime(df["event_time"], utc=True)
df["hour"]       = df["event_time"].dt.hour

hora_counts = df.groupby("hour").size()
hours       = hora_counts.index.tolist()
values      = hora_counts.values
 
max_hora = hora_counts.idxmax()
min_hora = hora_counts.idxmin()
max_val  = hora_counts[max_hora]
min_val  = hora_counts[min_hora]
mean_val = hora_counts.mean()
```

```python
# PALETA: color por intensidad relativa
# ─────────────────────────────────────────────
norm   = plt.Normalize(vmin=values.min(), vmax=values.max())
cmap   = plt.cm.Blues
colors = [cmap(norm(v)) for v in values]
# Resaltar pico y mínimo con colores propios
colors[max_hora] = "#1D4ED8"   # azul oscuro — máximo
colors[min_hora] = "#FCA5A5"   # rojo suave  — mínimo
```

```python
# FIGURA
fig, ax = plt.subplots(figsize=(14, 6))
fig.patch.set_facecolor("#F8FAFC")
ax.set_facecolor("#F8FAFC")
 
bars = ax.bar(hours, values, color=colors, width=0.7,
              edgecolor="white", linewidth=0.6, zorder=3)
 
# Línea de media
ax.axhline(mean_val, color="#64748B", linewidth=1.2,
           linestyle="--", zorder=2)
ax.text(23.55, mean_val, f"Media\n{int(mean_val):,}",
        fontsize=7.5, color="#64748B", va="center")
 
# Etiquetas en TODAS las barras (diagonal)
for bar, val, hora in zip(bars, values, hours):
    es_max = (hora == max_hora)
    es_min = (hora == min_hora)
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + max_val * 0.008,
        f"{int(val):,}",
        ha="left", va="bottom",
        fontsize=7 if (es_max or es_min) else 6.5,
        color="#1D4ED8" if es_max else ("#DC2626" if es_min else "#334155"),
        fontweight="bold" if (es_max or es_min) else "normal",
        rotation=45
    )
 
# Ejes
ax.set_xticks(range(24))
ax.set_xticklabels([f"{h:02d}:00" for h in range(24)],
                   rotation=45, ha="right", fontsize=8.5)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(
    lambda x, _: f"{int(x):,}"))
ax.tick_params(axis="both", length=0)
ax.set_ylabel("Número de Interacciones", fontsize=10, color="#334155", labelpad=10)
ax.set_xlabel("Hora del Día (UTC)", fontsize=10, color="#334155", labelpad=10)
 
# Cuadrícula horizontal suave
for y in ax.get_yticks():
    ax.axhline(y, color="#E2E8F0", linewidth=0.7, zorder=1)
 
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_color("#E2E8F0")
ax.spines["bottom"].set_color("#E2E8F0")
 
# Títulos
ax.set_title("Interacciones por Hora del Día",
             fontsize=13, fontweight="bold", loc="left",
             pad=28, color="#0F172A")
ax.text(0.0, 1.045,
        "Volumen agregado de eventos · Datos: 2020-04-01 → 2020-04-30 (UTC)",
        transform=ax.transAxes, fontsize=8.5, color="#64748B")
 
# Estadísticas resumen (esquina)
stats_text = (
    f"n total = {int(hora_counts.sum()):,}\n"
    f"Media/hora = {int(mean_val):,}"
)
ax.text(0.99, 0.97, stats_text,
        transform=ax.transAxes,
        fontsize=8, color="#475569",
        va="top", ha="right",
        bbox=dict(boxstyle="round,pad=0.4", fc="#F1F5F9", ec="#CBD5E1", lw=1))
 
plt.tight_layout()
plt.savefig("outputs/03_temporal_hourly_patterns.png", dpi=150,
            bbox_inches="tight", facecolor=fig.get_facecolor())
plt.close()
print("¡Gráfico 03_temporal_hourly_patterns.png guardado exitosamente!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Calculamos los porcentajes originales antes de limpiar el dataset
df_original = pd.read_csv("../../data/raw/2020-Apr.csv", usecols=["event_type"])
pct_originales = (df_original["event_type"].value_counts(normalize=True) * 100).to_dict()
```

```python
df_limpio = pd.read_csv("../../data/raw/2020-Apr-L.csv")

df_counts = df_limpio['event_type'].value_counts().reset_index()
df_counts.columns = ['event_type', 'conteo_absoluto']
total_limpio = df_counts['conteo_absoluto'].sum()
df_counts['porcentaje'] = (df_counts['conteo_absoluto'] / total_limpio) * 100
```

```python
# CONFIGURACIÓN DEL GRÁFICO
# Usamos un estilo más limpio y profesional
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 7))

colores = {'view': '#2461bd', 'cart': '#f97316', 'purchase': '#16a34a'}
categorias_ordenadas = ['view', 'cart', 'purchase']

ax = sns.barplot(
    x='porcentaje', 
    y='event_type', 
    data=df_counts, 
    palette=colores,
    order=categorias_ordenadas,
    alpha=0.9
)

# Iteramos usando explícitamente las posiciones del eje Y
for posicion_y, evento in enumerate(categorias_ordenadas):
    # 1. Conseguir datos del DataFrame Limpio
    fila_limpia = df_counts[df_counts['event_type'] == evento].iloc[0]
    conteo_limpio = fila_limpia['conteo_absoluto']
    pct_limpio = fila_limpia['porcentaje']
    
    # 2. Conseguir porcentaje Pre-limpieza
    pct_pre = pct_originales.get(evento, 0)
    
    # DIBUJAR LÍNEA VERTICAL PRE-LIMPIEZA
    ax.vlines(
        x=pct_pre, 
        ymin=posicion_y - 0.28, 
        ymax=posicion_y + 0.28, 
        color='#dc2626', # Rojo más elegante
        linestyle='--', 
        linewidth=2,
        label='Pre-limpieza (Crudo)' if posicion_y == 0 else ""
    )
    
    # ETIQUETA DOBLE (Al final de la barra sólida)
    texto_barra = f"{conteo_limpio:,} ({pct_limpio:.2f}%)"
    
    # Ponemos el texto un poco a la derecha del fin de la barra limpia
    ax.text(
        x=pct_limpio + 2, 
        y=posicion_y, 
        s=texto_barra, 
        color='#1f2937', 
        va='center', 
        ha='left',
        fontsize=10.5, 
        fontweight='semibold'
    )
    
    # ETIQUETA DE REFERENCIA PRE-LIMPIEZA
    # Se coloca flotando cerca de la línea punteada roja
    desplazamiento_x = 1.2 if pct_pre < 85 else -1.2
    alineacion_h = 'left' if pct_pre < 85 else 'right'
    
    ax.text(
        x=pct_pre + desplazamiento_x, 
        y=posicion_y - 0.15, # Ligeramente desplazado hacia arriba de la barra
        s=f"Pre: {pct_pre:.2f}%", 
        color='#dc2626', 
        va='center', 
        ha=alineacion_h,
        fontsize=9.5, 
        fontweight='bold',
        bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="#dc2626", lw=0.5, alpha=0.8) # Un pequeño globo de texto
    )

# DETALLES ESTÉTICOS FINALES
plt.title("Distribución de Eventos: Limpio vs Pre-Limpieza", fontsize=15, fontweight='bold', pad=20)
plt.xlabel("Porcentaje (%)", fontsize=12, labelpad=10)
plt.ylabel("Tipo de Evento", fontsize=12, labelpad=10)

# Le damos un margen extra al eje X para que las etiquetas largas no se corten
plt.xlim(0, df_counts['porcentaje'].max() + 20) 

# Mejoramos la leyenda
plt.legend(loc='lower right')
#plt.legend(loc='lower right', frameon=True, facecolor='white', edgecolor='none')
#sns.despine(left=True, bottom=True) # Quita los bordes innecesarios
plt.tight_layout()

#plt.show()

# guardar grafico
plt.savefig(
     "outputs/01_dist_event_type.png", 
    dpi=300,            # Alta resolución (ideal para tu documento/tesis)
    bbox_inches='tight' # Evita que las etiquetas externas se corten al guardar
)
plt.close()
print("¡Gráfico 01_dist_event_types.png guardado exitosamente!")
```

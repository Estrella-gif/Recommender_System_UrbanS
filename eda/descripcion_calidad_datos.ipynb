```python
import pandas as pd
df = pd.read_csv("../data/2020-Apr-L.csv")
```

```python
print("SHAPE:")
print(df.shape)

print("\nCOLUMNAS Y TIPOS:")
print(df.dtypes)
```

```python
print("VALORES NULOS:")
print(df.isnull().sum())
```

```python
print("DUPLICADOS EXACTOS(4 campos):")
dupesParcial = df.duplicated(subset=['event_time','event_type','product_id','user_id']).sum()
print(f"Duplicados: {dupesParcial:,}")

print("\nDUPLICADOS EXACTOS (Todos los campos):")
dupestotal = df.duplicated().sum()
print(f"Duplicados: {dupestotal:,}")
```
